```python
# Import Statements
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import in the CSV file
df = pd.read_csv('/home/data_scientist/checkin_dataset.csv')
df = df.drop('Unnamed: 0',axis=1) # Delete the unnamed column
df.head(10)   # check to see first 10 rows
```

```python
# With this cell I am creating a new DataFrame that allows me to see all of 
# the median or 50th percentile values for each users x and y coordinates
d = []
for i in range(0,100):
    x_describe = df[df.user == i]['xcoordinate'].describe()
    y_describe = df[df.user == i]['ycoordinate'].describe()
    x_med = x_describe[5]
    y_med = y_describe[5]
    d.append((i,x_med,y_med))

df_update = pd.DataFrame(d,columns=('user','xcoordinate median','ycoordinate median'))
display(df_update)
```

From the given data above we can determine what x coordinate and what y coordinate the the user is most often at using the medians found for the x and y. What this means is that we are able to see not what locations they are most generally at. 

```python
# What this cell allows us to do is find the mode of the data that we just
# generated
temp1 = round(df_update['xcoordinate median'].mode(),4)
temp2 = round(df_update['ycoordinate median'].mode(),4)
x_mode = round(temp1.value_counts().idxmax(),4)
y_mode = round(temp2.value_counts().idxmax(),4)
display(x_mode)
display(y_mode)
```
